# Standard Brownian Motion Sample Paths

## Here we create a standard Brownian motion realization on an interval [0,T] starting at 0.
## This starts at 0 and makes jumps with mean 0 and variance $\sqrt{dt}$ in every time interval of length $dt=T/N$.

## We create a sequence of jumps 

> ## $\delta_1,\delta_2,\ldots,\delta_{N}$

## then compute the cumulative sums (including the starting value of 0)

> ## $0,\delta_1,\delta_1+\delta_2,\delta_1+\delta_2+\delta_3,\ldots,\delta_1+\delta_2+\delta_3+\cdots+\delta_N.$

## So that the points on the graph of the resulting random function are:

> ## $(0,0),(dt,\delta_1),(2dt,\delta_1+\delta_2),\ldots,(T=Ndt,\delta_1+\delta_2+\delta_3+\cdots+\delta_N)$

```python
import plotly.graph_objects as go
import numpy as np
import math

T=1
N=1000
dt=T/N
sqrtdt = math.sqrt(dt)
dB=np.random.normal(0,sqrtdt,N)
B=np.cumsum(dB)
B=np.insert(B,0,0) # insert 0 at begginning of B
T=np.linspace(0,1,N+1)
```

# Geometric Brownian Motion

## Geometric Brownian motion (GBM)is easy to understand once you understand Brownian motion. Standard GBM is obtained by making an exponential transformation of Brownian motion. In other words, if B(t) is standard Brownian motion, then $S(t) = e^{B(t)}$ gives us geometric Brownian motion. One important property of this stochastic process is that it never becomes negative.

```python
T=1
N=1000
dt=T/N
sqrtdt = math.sqrt(dt)
dB=np.random.normal(0,sqrtdt,N)
B=np.cumsum(dB)
B=np.insert(B,0,0) # insert 0 at beginning of B
S=np.exp(B)
T=np.linspace(0,1,N+1)
```

# A function for generating Brownian motion realizations

```python
import numpy as np
import math
#
# Inputs: 
#
#   T0 = initial time
#   T1 = final time
#   B0 = starting value (B(T0))
#   Ninc = number of increments
#
def BrownianMotion(T0,T1,B0, Ninc):
    dt=(T1-T0)/Ninc
    sqrtdt=math.sqrt(dt)
    dB=np.random.normal(0,sqrtdt,Ninc)
    B=np.cumsum(dB)
    B=np.insert(B,0,0)
    B=B0+B
    T=np.linspace(T0,T1,Ninc+1)
    return((T,B))
    
```

```python
T0=3.2
T1=18.5
B0=7.8
Ninc=1000
T,B1=BrownianMotion(T0,T1,B0,Ninc)
T,B2=BrownianMotion(T0,T1,B0,Ninc)
T,B3=BrownianMotion(T0,T1,B0,Ninc)
```

```python
import plotly.graph_objects as go

# Create random data with numpy
import numpy as np
np.random.seed(1)

# Create traces
fig = go.Figure()
fig.add_trace(go.Scatter(x=T, y=B1,
                    line=dict(color='red', width=.5),
                    mode='lines',
                    name='realization 1'))
fig.add_trace(go.Scatter(x=T, y=B2,
                    line=dict(color='blue', width=.5),
                    mode='lines',
                    name='realization 2'))
fig.add_trace(go.Scatter(x=T, y=B3,
                    line=dict(color='green', width=.5),
                    mode='lines', 
                    name='realization 3'))

#fig.update_layout(title_text="A Bar Chart",
#                  title_font_size=20)
fig.update_layout(title=dict(text="Multiple Brownian Motion Realizations",x=.5,font_size=20))
fig.update_layout(xaxis=dict(title="t",title_font_size=15))
fig.update_layout(yaxis=dict(title="B(t)",title_font_size=15))
fig.update_layout(showlegend=False)
fig.show()
```

# What is the chance that a standard Brownian motion path starting from 7.8 at time 3.2 goes above 10 any time before 18.5?


```python
import numpy as np
import math
#
# Inputs: 
#
#   T0 = initial time
#   T1 = final time
#   B0 = starting value (B(T0))
#   Ninc = number of increments
#
def BrownianMotion(T0,T1,B0, Ninc):
    dt=(T1-T0)/Ninc
    sqrtdt=math.sqrt(dt)
    dB=np.random.normal(0,sqrtdt,Ninc)
    B=np.cumsum(dB)
    B=np.insert(B,0,0)
    B=B0+B
    T=np.linspace(T0,T1,Ninc+1)
    return((T,B))


T0=3.2
T1=18.5
B0=7.8
Ninc=10
N=10000
ctr=0
for i in range(N):
    T,B=BrownianMotion(T0,T1,B0,Ninc)
    if np.max(B)>10.:
        ctr=ctr+1
pest=ctr/N
sd=math.sqrt(pest*(1-pest)/N)
print("Estimated probability = "+ str(pest))
print("Standard error = " + str(sd))
```
